```python
# -*- coding: utf-8 -*- 
import tushare as ts
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
```

```python
HeaderList = ['600','601','603','000','002']
FixedNumber = '85'
n = range(10)
MyList = []

for head in HeaderList:
    for i in n:
        MyList.append(head+str(i)+FixedNumber)

rd = ts.get_report_data(2016,3)
myrd = rd[rd.code.isin(MyList)]
ERD1603 = myrd[(myrd.eps>0.1)&(myrd.roe>5)].drop_duplicates()
List = ERD1603.code.values
```

## 结合上市公司基本情况根据经验选择中国核电（601985）

```python
#上市公司基本情况--sb
a = ts.get_stock_basics()
sb = a.loc[List]
erd = ERD1603.ix[ : ,['code','roe']]
erd = erd.set_index('code')
SB = sb[['name','timeToMarket','outstanding','bvps','holders','industry','area','esp']]
SB = SB.merge(erd,left_index=True,right_index=True)
SB = SB.sort_values(by='outstanding',ascending=False)
SB.columns = ['名称','上市日期','流通股本（亿）','每股净资','股东人数','所属行业','地区','EPS','ROE']
SB
```
